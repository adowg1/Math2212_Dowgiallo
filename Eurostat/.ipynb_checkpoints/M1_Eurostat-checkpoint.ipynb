## ADDING ROWS TO A DATAFRAME

The ideam is to make the collection of new rows into a new dataframe. Then you just "glue" that new dataframe onto the end of your original dataframe

```python
# to create a new dataframe, start with a directory
new_rows_dict = {
    'TIME': [2000,2000,2000],
    'GEO' : ['USA', 'Canada', 'Uganda'],
    'Value' : [5, 4.18, 5.15]
}

#turns the dictionary into a dataframe 
new_rows_df = pd.DataFrame(new_rows_dict)

new_rows_df
```

```python
# Append the new_rows_df to the end of edu
edu_new_rows = edu.append(new_rows_df, ignore_index=True, sort=False)

edu_new_rows.tail()
```

```python
# Worksheet 4
ws4_rows_dict = {
    'TIME': [2020,2020,2020],
    'GEO' : ['USA', 'Germany', 'Mexico'],
    'Value' : [55, 4.32, 94]
}

ws4_rows_df = pd.DataFrame(ws4_rows_dict)

edu_ws4_rows = edu_new_rows.append(ws4_rows_df, ignore_index=True, sort=False)

edu_ws4_rows.tail()
```

# Filtering Dataframes
Boolean filtering

```python
# Return the rows with value bigger than 4
edu[edu['Value'] > 4]
```

```python
# Return the rows with value between 4 and 6
edu[ (edu['Value'] > 4) & (edu['Value'] < 6) ]
```

```python
# Return the rows with years 2000 and 2003
edu[ (edu['TIME'] == 2002) | (edu['TIME'] == 2003) ]
```

```python
# Filter some of the columns 
edu[edu['TIME'] == 2002 ] [["GEO" , "Value"]]
```

```python
# Return the rows with value between 5 and 7
edu[ (edu['Value'] > 5) & (edu['Value'] < 7) ]
```

```python
# Filter some of the columns 
edu[edu['TIME'] == 2003 ] [["GEO"]]
```

```python
# Return on the rows with missing Value data
#edu[ edu['Value'].isnull() ]

# Return the rows with no missing data
edu[ (edu['Value'].notnull()) |(edu["GEO"].notnull())]
```

# Aggregating Data
how can we sum, average, or otherwise aggregate data that shares a similiar property?

```python
group_mean = edu[['GEO', 'Value']].groupby('GEO').mean() # finds the mean value for each country
print group_mean
```

```python
group_max = edu[['TIME', 'Value']].groupby('TIME').sum()
print group_max
```

# Pivot Tables
A pivot table is similar to these tables that we got using .groupby() except that instead of aggregating over one variable, we aggregate over two variables.

```python
ptable = pd.pivot_table( 
    edu,
    values = 'Value',
    index = 'TIME',
    columns = 'GEO',
    aggfunc = np.sum
)
ptable
```

```python
#WS #6
ptable6 = pd.pivot_table( 
    edu,
    values = 'Value',
    index = 'GEO',
    columns = 'TIME',
    aggfunc = np.mean
)
ptable6
```

# Plotting Data

The basic format to plot data is:

dataframe.plot(x ="Column1", y ="Column2, kind = GraphType, title = 'Title')

```python
# Scatter Plot
edu.plot(x= 'TIME', y = 'Value', kind = 'scatter', title = 'Title')
```

```python
# Line Plot

edu_austria = edu[ edu['GEO'] == 'Austria']

edu_austria.plot(x = 'TIME', y = 'Value', kind = 'line', title = 'Austria')
```

```python
# Bar Graph
# A bar graph of the total sum value within each geo region

edu_country_value_aggregate = edu[['GEO','Value']].groupby('GEO').sum()

edu_country_value_aggregate.plot(kind = 'bar')
```

```python
# Bar Graph
# A bar graph of the total sum value within each geo region

edu_country_value_aggregate = edu[['GEO','Value']].groupby('GEO').sum()

edu_country_value_aggregate.plot(kind = 'barh')
```
