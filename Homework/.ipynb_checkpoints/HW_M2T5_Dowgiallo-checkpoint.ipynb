```python
import  pandas  as pd
import  numpy  as np
import  matplotlib.pyplot  as plt
```

```python
import seaborn as sns
```

```python
loans = pd.read_csv('loans_full.csv')

pd.set_option('display.max_columns', None)

loans.head()
```

# Question 1

```python
loans_corr_matrix = loans.corr()
print loans_corr_matrix
```

```python
sns.heatmap(
    loans_corr_matrix,
    xticklabels = loans_corr_matrix.columns,
    yticklabels = loans_corr_matrix.columns,
    annot = True
)
```
