# Question 1
Find means, standard deviations, medians, and interquartile ranges

```python
# Annual Income Data
print loans['annual_income'].mean()
print loans['annual_income'].std()
print loans['annual_income'].median()
IQR = loans['annual_income'].quantile(0.75) -  loans['annual_income'].quantile(0.25)
print IQR
```

```python
# Debt-to-income ratio
print loans['debt_to_income'].mean()
print loans['debt_to_income'].std()
print loans['debt_to_income'].median()
IQR = loans['debt_to_income'].quantile(0.75) -  loans['debt_to_income'].quantile(0.25)
print IQR
```

```python
# The difference between the total credit available and credit used
difference = loans['total_credit_limit'] - loans['total_credit_utilized']
print difference.mean()
print difference.std()
print difference.median()
IQR = difference.quantile(0.75) -  difference.quantile(0.25)
print IQR
```

```python
# Interest Rate
print loans['interest_rate'].mean()
print loans['interest_rate'].std()
print loans['interest_rate'].median()
IQR = loans['interest_rate'].quantile(0.75) -  loans['interest_rate'].quantile(0.25)
print IQR
```

# Question 5

Remove some outliers from the debt-to-income ratio using the mean method and again using the median methods. Draw boxplots before and after removing the outliers.

```python
loans.boxplot('debt_to_income_joint')
```

```python
debt = loans['debt_to_income_joint']

debt_mean = debt.mean()
debt_std = debt.std()

loans_mean_outlier_drop = loans.drop(
    loans.index[
        (debt > debt_mean + 2.5*debt_std) | (debt < debt_mean - 2.5*debt_std)
]
)

print debt_mean - 2.5*debt_std
print debt_mean + 2.5*debt_std

loans_mean_outlier_drop.head()
```

```python
loans_mean_outlier_drop.boxplot()
```

```python
debt = loans['debt_to_income_joint']

debt_median = debt.median()
debt_iqr = debt.quantile(0.75) - debt.quantile(0.25)

loans_median_outlier_drop = loans.drop(
    loans.index[
        (debt > debt_median + 2.5*debt_iqr) | (debt < debt_median - 2.5*debt_iqr)
]
)

print debt_median - 2.5*debt_iqr
print debt_median + 2.5*debt_iqr

loans_median_outlier_drop.head()
```

```python
loans_median_outlier_drop.boxplot()
```
